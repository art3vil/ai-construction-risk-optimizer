```python
import pandas as pd
import matplotlib.pyplot as plt

from src.train_regressor import FEATURE_COLUMNS, _encode_categories
import joblib

# Загрузка данных и модели риска
df = pd.read_csv("data/raw/synthetic_construction_projects.csv")
clf = joblib.load("models/risk_model.pkl")

# Подготовка признаков
enc = _encode_categories(df)
X = enc[FEATURE_COLUMNS]

# Предсказанная вероятность перерасхода для всех проектов
df["risk_pred"] = clf.predict_proba(X)[:, 1]

# Средний риск по классам районов
risk_by_district = df.groupby("district_class")["risk_pred"].mean()

plt.figure()
risk_by_district.plot(kind="bar")
plt.title("Средний риск перерасхода по классам районов")
plt.ylabel("Вероятность перерасхода")
plt.show()

# Средний риск по сезонам
plt.figure()
df.groupby("weather_season")["risk_pred"].mean().plot(kind="bar")
plt.title("Средний риск по сезонам")
plt.ylabel("Вероятность перерасхода")
plt.show()

# Средний риск по типу клиента
plt.figure()
df.groupby("client_type")["risk_pred"].mean().plot(kind="bar")
plt.title("Средний риск по типу клиента")
plt.ylabel("Вероятность перерасхода")
plt.show()
```
